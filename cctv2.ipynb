```python
import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow

# Upload the input image file
uploaded = files.upload()

# Read the uploaded image
image_path = next(iter(uploaded))
image = cv2.imdecode(np.frombuffer(uploaded[image_path], np.uint8), cv2.IMREAD_COLOR)

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply histogram equalization to improve contrast
equ = cv2.equalizeHist(gray)

# Apply adaptive histogram equalization for local contrast enhancement
clahe = cv2.createCLAHE(clipLimit=5.0, tileGridSize=(8, 8))
clahe_image = clahe.apply(gray)

# Combine the equalized image and the CLAHE image to produce the enhanced image
enhanced_image = np.hstack((equ, clahe_image))

# Display the enhanced image
cv2_imshow(enhanced_image)

# Save the enhanced image
enhanced_image_path = f'enhanced-{image_path}'
cv2.imwrite(enhanced_image_path, enhanced_image)
```
